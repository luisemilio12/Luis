```python
#np.arange(start,end,step)
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
```

```python
ysin=np.sin(x)
ycos=np.cos(x)

plt.plot(x,ysin)
plt.plot(x,ycos)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('Sine & Cosine functions')
```

```python
plt.subplot(2,1,1)
plt.plot(x,ysin)
plt.title('sine')
plt.subplot(2,1,2)
plt.plot(x,ycos)
plt.title('Cosine')
```

```python
path ='datasets/Cartwheeldata.csv'
df=pd.read_csv(path)
df.head(2)
```

```python
sns.lmplot(x='Wingspan',
           y='Height',
           data=df,
           fit_reg=False,
           hue= 'Gender'
          )
plt.title('Scatter plot of Wingspan vs Height')
```

```python
sns.swarmplot(x='Gender',
              y='Wingspan',
              data=df,
             )
plt.title('Gender vs Wingspan, the swarmplot')
```

```python
sns.boxplot(data=df.loc[df['Gender']=='M',
                        ['Age','Height','Wingspan','CWDistance']])
plt.title('male Boxplot')
```
