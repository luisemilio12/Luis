```python
import statsmodels.api as sm
```

```python
df= pd.read_csv('nhanes_2015_2016.csv')
df.columns
```

```python
vars = ['SMQ020', 'RIAGENDR', 'RIDAGEYR','BPXSY1','BMXBMI', 'BMXLEG']
df = df[vars].dropna()
df
```

```python
df[vars].corr()
```

```python
model = sm.OLS.from_formula ('BPXSY1 ~ RIAGENDR',
                            data = df)
result = model.fit()
result.summary()
```

```python
ax = sns.regplot (x = 'BPXSY1',
                  y = 'RIAGENDR',
                  data = df,
                 x_estimator = np.mean,
                 )
```
