```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%ls datasets/
```

```python
df= pd.read_csv('datasets/antropometria-dataset (1).csv')
df.head()
```

```python
model = sm.OLS.from_formula ('cadera ~ peso',
                            data = df)
result = model.fit()
result.summary()
```

```python
vars = ['folio', 'intp', 'entidad', 'desc_ent', 'sexo', 'edad', 'meses', 'peso',
       'ropa', 'talla', 'emb', 'temb', 'cintura', 'cadera', 'sistol',
       'diastol', 'hpresion', 'tbrazo', 'htension', 'PrimaryLast', 'code_upm',
       'est_dis', 'est_urb', 'est_marg', 'pondef', 'est_var']
df = df[vars].dropna()
df[vars].corr()
```

```python
model = sm.OLS.from_formula ('diastol ~ sistol',
                            data = df)
result = model.fit()
result.summary()
```

```python
model2 = sm.OLS.from_formula ('diastol ~ sistol+ edad + peso + cintura',
                            data = df)
result = model2.fit()
result.summary()
```

```python
y=result.resid
y.median()
```
