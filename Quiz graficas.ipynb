```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df
```

```python
sns.boxplot(data=df.loc[:,['edad']])
```

```python
tabla = pd.pivot_table(df,
            index = ['desc_ent'],
            columns = ['sexo'],
            values = ['peso'],
            aggfunc = {'peso':['median']}
              )
tabla
```

```python
tabla.columns
```

```python
boxplot = tabla.boxplot(column=[('peso', 'median', 1)])
```

```python
tablados = pd.pivot_table(df,
            index = ['desc_ent'],
            columns = ['sexo'],
            values = ['cintura'],
            aggfunc = {'cintura':['median']}
              )
tablados
```

```python
line_plot = tablados[('cintura','median',2)].plot(kind='line')
```

```python
tablatres = pd.pivot_table(df,
            index = ['desc_ent'],
            columns = ['sexo'],
            values = ['ropa'],
            aggfunc = {'ropa':['sum']}
              )
tablatres
```

```python
by_ropa = tablatres.sort_values(('ropa','sum',2),ascending=False)
top10 = by_ropa.head(10)
top10
```

```python
top10[('ropa','sum',2)].plot(kind = 'barh',
            figsize = (20,10),
            color = 'y',
            alpha = 0.5)
```
