```python
df= pd.read_csv('datasets/sample_pivot.csv')
df.head(2)
```

```python
average_sales = pd.pivot_table(df,
                               index ='Region',
                               values = 'Sales',
                               aggfunc=['sum','mean','median'],
                               columns='Type',
                               margins='sum'
                              )
average_sales
```
