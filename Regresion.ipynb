### 1. existe alguna relacion entre 'Brain Weight y Head size(cm^3)'

```python
df=pd.read_csv('datasets/headbrain.csv')

x=df['Head Size(cm^3)']
y=df['Brain Weight(grams)']

np.corrcoef(x,y)
```

#### Conclusion
El coeficiente de correlacion es de .799, por lo que existe una alta correlacion entre Brain Weight y Head size


```python
colors={1:'red', 2:'green'}
plt.figure(figsize=(13,8))
plt.scatter(x,y,
           c=df['Gender'].map(colors),
           marker= 'x',
           s=175)
plt.xlabel('Head size (cm^3)', fontweight = 'regular', color= 'gold', fontsize=18, horizontalalignment='center')
plt.ylabel('Brain Weight(grams)',  fontweight = 'regular', color= 'blue', fontsize=18, horizontalalignment='center')
plt.title('Scatterplot de brain weight vs head size')
```
